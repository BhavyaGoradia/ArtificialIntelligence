```python

df1 = df[["reviews.text","reviews.rating"]]
```

```python

corpus = df1["reviews.text"].tolist()
corpus
```

```python
from nltk.corpus import stopwords
from string import punctuation
stuff_to_be_removed = list(stopwords.words("english"))+list(punctuation)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.lancaster import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
stemmer = LancasterStemmer()
```

```python
final_corpus= []
for i in range(len(corpus)):
    text = word_tokenize(corpus[i].lower())
    text = [lancaster_stemmer.stem(y) for y in text if y not in stuff_to_be_removed]
    sent = " ".join(text)
    
    final_corpus.append(sent)
```

```python
new_df = pd.DataFrame(final_corpus)
```

```python
new_df["reviews.rating"] = df1["reviews.rating"]
```

```python
new_df.head()
```

```python
new_df.columns = ["reviews.text", "reviews.rating"]
```

```python
final_corpus1= []
for i in range(len(final_corpus)):
    text = word_tokenize(final_corpus[i])
    tagged = nltk.pos_tag(text)
    adjective_tags = ["JJ", "JJR", "JJS"]
    close_adjectives_list = [a[0] for a in tagged if a[1] in adjective_tags]
    sent = " ".join(text)
    final_corpus1.append(sent)
```

```python
final_corpus1
```

```python
list_demo=[]
for i in range (len(final_corpus)):
    counts = dict()
    words = final_corpus[i].split(" ")
    
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    list_demo.append(counts)
```

```python
df = pd.DataFrame(pd.Series(list_demo).reset_index()).set_axis(['Key','Value'],1,inplace=False)
```
